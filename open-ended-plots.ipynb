```python
import pandas as pd 
import os 
import numpy as np
import json
import re
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


# Loop through files in other folder
# Read csv to df
# if ['category'] is non-empty: 
    # plot value counts bar graph

open_dir = os.path.join('data', 'manual_categories', 'open')
other_dir = os.path.join('data', 'manual_categories', 'other')

def plot_bar(data, title,text_buffer=.05, bar_color='skyblue', color_map=None, text_counts=True, display_data=True, legend_map=None): 

    plt.figure(figsize=(10, 6))
    bars = plt.barh(data.index, data.values, color=bar_color)
    plt.title(title)
    plt.gca().invert_yaxis()  # Invert y-axis to have the highest value at the top
    # Add text after the bars to show count 

    # Adding count annotation to each bar
    if text_counts: 
        for bar in bars:
            plt.text(bar.get_width() + text_buffer, bar.get_y() + bar.get_height()/2, 
                        f'{int(bar.get_width())}', 
                        va='center', ha='left', fontsize=10, color='black')
            
    if color_map is not None: 
        # display(data)
        for i, bar in enumerate(bars):
            category = data.index[i]
            if category in color_map:
                bar.set_color(color_map[category])
            
   
    if legend_map is not None: 
        legend_handles = [mpatches.Patch(color=color, label=label) for label, color in legend_map.items()]
        plt.legend(handles=legend_handles)
                
    plt.show()

    if display_data: # Note is ignored if not showing the plot, too 
        data_to_display = data.reset_index()
        data_to_display.columns = [col.title().replace('_', ' ') for col in data_to_display.columns]
        data_to_display.index = list(range(1,data.shape[0]+1))
        print(f'Count Column: {data_to_display["Count"].sum()}')
        display(data_to_display)



#### Open Ended
to_exclude = []
# to_exclude = ['How do you think the training or education program could have been better?',
#             #   'Where did you participate in a workforce or employment training program?',
#               'What is your ideal job?', 
#               'What skills would you most like to develop?'
#               'What are your current job title(s)']
            
non_assigned = []
color_map = None
legend_map=None
for fp in os.listdir(open_dir):
    print(fp)
    df = pd.read_csv(os.path.join(open_dir, fp), index_col=0)
 
    ## Change this to only print a specific chart or chart 
    # if 'barriers' not in fp:
    #     continue
 
    if df['category'].isna().sum() != df.shape[0]:
        question = fp.split('_')[0]
        if question in to_exclude:
            continue
        # print(question)

        n_non_nan = df.shape[0] - df[question + '_Open'].isna().sum()
        # n_non_nan = df[~df[question + '_Open'].isna()].shape[0]

        # Set color maps for other questions
        if question == 'Where did you participate in a workforce or employment training program?':
            color_map = {'Other States':'lightsteelblue', 'Delaware':'lightsteelblue'}
            
        # Split ideal job question between industry and "work perks" (job criteria)
        if question == 'What is your ideal job?':

            legend_map = {'Role or Industry Type':'skyblue', 'General Job Criteria':'lightsteelblue'}

            color_map = {'General Job Criteria':'lightsteelblue', 'On Disability':'lightgrey'}

            # Industry/Position
            if os.path.basename(fp) == 'What is your ideal job?_n=261.csv':

                industries = df[df['category'] != 'General Job Criteria']
                job_criteria = df[df['category'] == 'General Job Criteria']

                # v_counts = industries['category'].value_counts()

                v_counts = df['category'].value_counts()
                plot_bar(v_counts, question + f" (N={n_non_nan})", 
                         bar_color='skyblue', color_map=color_map, display_data=False, legend_map=legend_map) 

                data_to_display = industries['category'].value_counts().reset_index().rename({"category":"Industry/Position",
                                                                                              "count":"Count"}, axis=1)
                data_to_display.index = [x + 1 for x in data_to_display.index]
                display(data_to_display)

                display('Count', data_to_display['Count'].sum())

            # Job Criteria
            elif os.path.basename(fp) == 'What is your ideal job?_n=261_(General Job Criteria).csv':

                v_counts = df['job_criteria'].value_counts()
                plot_bar(v_counts, bar_color='lightsteelblue', title=question + f" (N={n_non_nan}) (Job Criteria)", 
                         display_data=True)
                
        elif question == 'How do you think the training or education program could have been better?':
            
            color_map = {'More Specialized Content':'lightsteelblue'}
            v_counts = df['category'].value_counts()
            plot_bar(v_counts, bar_color='skyblue', title=question + f" (N={n_non_nan})", 
                        display_data=True,color_map=color_map)
            
            data_to_display = df[df['category'] == 'More Specialized Content']\
                [['How do you think the training or education program could have been better?_Open']]\
                .reset_index(drop=True)
            
            data_to_display.index = [x + 1 for x in data_to_display.index]

            display(df)
          
            for line in data_to_display.to_csv(index=False).splitlines():
                print(line.replace(',', ', '))

        elif question == 'Why havent you applied for unemployment benefits?':
            title = question + f" (N={n_non_nan - 1})"
            v_counts = df['category'].value_counts()
            plot_bar(v_counts, title=title, color_map=color_map)
    
        else: 
            print(legend_map)
            v_counts = df['category'].value_counts()
            if "?" not in question:
                plot_bar(v_counts, title=question + f"? (N={n_non_nan})", color_map=color_map, legend_map={'Role or Industry Type':'skyblue'})
            else: 
                plot_bar(v_counts, title=question + f" (N={n_non_nan})", color_map=color_map, legend_map=legend_map)    

          
    else: 
        non_assigned.append(fp)

if len(non_assigned) != 0: 
    display(non_assigned)


    # display(v_counts)
```

```python
# MC Answers 
df_main = pd.read_csv(os.path.join('data','results','DEEP_Cleaned_Survey_Responses_S1.csv'), index_col=0)
df_main_barrier = df_main.filter(regex='barrier|Respondent ID')

# 'Other' Answer Categories
fp = 'What are some barriers that you feel you face to achieving your ideal job?_n=107.csv'
question = fp.split('_')[0]
barrier_other_categories = os.path.join(other_dir, fp)
df_barrier_other_categories = pd.read_csv(barrier_other_categories, index_col=0)

# Join and drop empty responses
df_barrier_joined = df_main_barrier.merge(df_barrier_other_categories, how='left', on=['Respondent ID', f'{question}_Other'])
numeric_cols = df_barrier_joined.columns[1:-2]
df_barrier_joined = df_barrier_joined[(~df_barrier_joined['category'].isna()) | (df_barrier_joined[numeric_cols].sum(axis=1) > 0)]
non_nan_count = df_barrier_joined.shape[0]
print(f'N Responses to Barrier Question: {non_nan_count}\n({question})')

df_barrier_joined['category'] = df_barrier_joined['category'].str.title()
 

display(df_barrier_joined)


# v_counts = df_barrier_other['category'].value_counts()
# non_nan_count = df_main[~df_main[question + "_Other"].isna()].shape[0]
# plot_bar(v_counts, title=f'What are some barriers that you feel you face to achieving your ideal job? (N={non_nan_count})')
# # v_counts
```

```python
df_barrier_combined = df_barrier_counts.merge(df_barrier_other_counts, on='category', how='outer')
df_barrier_combined['count'] = df_barrier_combined[['count_x', 'count_y']].sum(axis=1)
df_barrier_combined = pd.Series(df_barrier_combined['count'].values, index=df_barrier_combined['category']).sort_values(ascending=False)

df_barrier_combined.index = [x for x in df_barrier_combined.index]

df_barrier_combined
```

```python
other_cats = [x for x in df_barrier_other_counts['category'].values
                      if x not in df_barrier_counts['category'].values]
color_map = dict(zip(other_cats, ['lightsteelblue'] * len(other_cats)))
print('"Other" categories (disambiguated)\n' + json.dumps(color_map, indent=2))

legend_map = {'Given Categories':'skyblue', 'Free Responses (Categorized)':'lightsteelblue'}

plot = plot_bar(data=df_barrier_combined, title=f"{question} (N={non_nan_count})", 
         display_data=False, text_buffer=.7,
        bar_color='skyblue', color_map=color_map, legend_map=legend_map)
```

```python
import matplotlib.pyplot as plt

# Sample data
data = {
    "Category 1": [10, 15],
    "Category 2": [20, 25],
}

# Color map
color_map = {
    "Category 1": "r",
    "Category 2": "b"
}

# Extract labels and colors for legend
labels = list(color_map.keys())
colors = list(color_map.values())

# Plotting
fig, ax = plt.subplots()
index = range(len(data))
bar_width = 0.35
opacity = 0.8

for i, (label, color) in enumerate(color_map.items()):
    ax.bar([x + i * bar_width for x in index], data[label], bar_width, alpha=opacity, color=color, label=['label',';a'])

ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Title')
ax.set_xticks([i + bar_width for i in index])
ax.set_xticklabels(data.keys())
ax.legend()

plt.tight_layout()
plt.show()
```
